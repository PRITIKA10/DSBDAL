```python
titanic['sex'] = titanic['sex'].map({'male':0,'female':1})
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic['embarked'] = titanic['embarked'].map({'S':0,'C':1,'Q':2})
titanic['alive'] = titanic['alive'].map({'no':0,'yes':1})
titanic['alone'] = titanic['alone'].map({True:1, False:0})
```

```python
titanic.head()
```

```python
data = titanic.drop(columns = ['class','who','adult_male','deck','embark_town'])
```

```python
data.head()
```

```python
sns.heatmap(data);
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

```python
sns.histplot(data = titanic,x= 'age',bins = 30);
```

```python
sns.catplot(x ='embarked', hue ='survived', kind ='count', col ='class', data = titanic);
```

Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,

2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a box plot for each feature in the dataset.
4. Compare distributions and identify outliers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the iris CSV file into the pandas DataFrame
iris = pd.read_csv('Iris.csv')
iris
```

```python
iris.dtypes
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.histplot(data=iris, x='SepalLengthCm', bins=20, ax=axes[0, 0])
axes[0, 0].set_title('Histogram for Sepal Length')

sns.histplot(data=iris, x='SepalWidthCm', bins=20, ax=axes[0, 1])
axes[0, 1].set_title('Histogram for Sepal Width')

sns.histplot(data=iris, x='PetalLengthCm', bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Histogram for Petal Length')

sns.histplot(data=iris, x='PetalWidthCm', bins=20, ax=axes[1, 1])
axes[1, 1].set_title('Histogram for Petal Width')


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.boxplot(data=iris, x='Species', y='SepalLengthCm', ax=axes[0, 0])
axes[0, 0].set_title('Iris Sepal Length Boxplot')

sns.boxplot(data=iris, x='Species', y='SepalWidthCm', ax=axes[0, 1])
axes[0, 1].set_title('Iris Sepal Width Boxplot')

sns.boxplot(data=iris, x='Species', y='PetalLengthCm', ax=axes[1, 0])
axes[1, 0].set_title('Iris Petal Length Boxplot')

sns.boxplot(data=iris, x='Species', y='PetalWidthCm', ax=axes[1, 1])
axes[1, 1].set_title('Iris Petal Width Boxplot')

plt.tight_layout()
plt.show()
```
